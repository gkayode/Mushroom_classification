# Mushroom Classification

## Goal: To determine whether mushrooms are edible or poisonous based on their attribute

## Data Set Information:

The Dataset was gotten from the UCI Machine learning repository. https://archive.ics.uci.edu/ml/datasets/mushroom

This data set includes descriptions of hypothetical samples corresponding to 23 species of gilled mushrooms in the Agaricus and Lepiota Family (pp. 500-525). Each species is identified as definitely edible, definitely poisonous, or of unknown edibility and not recommended. This latter class was combined with the poisonous one. The Guide clearly states that there is no simple rule for determining the edibility of a mushroom; no rule like ''leaflets three, let it be'' for Poisonous Oak and Ivy.

# Importing Modules and Dataset

```python
#Importing modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Importing Dataset
column_labels = [
    'class', 'cap shape', 'cap surface', 'cap color', 'bruised', 'odor',
    'gill attachment', 'gill spacing', 'gill size', 'gill color', 
    'stalk shape', 'stalk root', 'stalk surface above ring',
    'stalk surface below ring', 'stalk color above ring',
    'stalk color below ring', 'veil type', 'veil color', 'ring number',
    'ring type', 'spore print color', 'population', 'habitat'
]
df=pd.read_csv('agaricus-lepiota.csv',names=column_labels)
df
```

## Data Cleaning

```python
df = df[df['stalk root'] != '?'] #drop missing data
df
```

# Label Encoding

```python
# This step basically involves representing the different attribute categories with numerical values
df_enc=pd.DataFrame({col: df[col].astype('category').cat.codes for col in df}, index=df.index)
df_enc
```

# Data Exploration and Visualization

```python
#Pandas profiling was used in exploring and visualizing the Data
# Please scroll through the result.
#Its also interactive
from pandas_profiling import ProfileReport
profile = ProfileReport(df.drop(columns=['veil type']), title="Pandas Profiling Report",html={'style':{'full_width':True}})
profile.to_notebook_iframe()
```

```python
# splitting data into predictors and target
X_df= df_enc.loc[:, df.columns != 'class']
Y_df=df_enc[['class']]
```

# Normalization

```python
# Standardize data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_std = scaler.fit_transform(X_df)
```

# Test-train

```python
#test-train split
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_std, Y_df,test_size=0.20)
```

# Cross-validation & Model fitting using SVC

```python
from sklearn.model_selection import RandomizedSearchCV
param_grid={"C": np.logspace(-1, 5, 10),
             "gamma": np.logspace(-5, 5, 10)}

# Create the model to be tuned
from sklearn.svm import SVC
n_iter = 2000 ### Setting low for now, so it will go fast.
folds = 4
svc_base = SVC(kernel='rbf')

# Create the random search 
svc_random = RandomizedSearchCV(estimator = svc_base, param_distributions = param_grid, scoring="neg_mean_squared_error",
                               n_iter = n_iter, cv = folds, verbose = 2, random_state = 42, 
                               n_jobs = -1)
```

```python
# Fit the random search model
svc_random.fit(X_train, Y_train)

# View the best parameters from the random search
print(svc_random.best_params_)

svc_model = svc_random.best_estimator_
svc_model.fit(X_train, Y_train)
```

# Model metrics

```python
# model metrics using f1_score
from sklearn.metrics import f1_score
Y_pred=svc_model.predict(X_test)
f1_score(Y_test, Y_pred)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
cm=confusion_matrix(Y_test,Y_pred,labels=[0,1])
confusion=pd.DataFrame(cm,index=['edible','poisonous'],columns=['predicted_edible','predicted_poisonous'])
confusion
```

```python
%matplotlib notebook
sns.heatmap(confusion,annot=True,fmt="d")
```

The model performs excellently, accurately predicting the test set. However it will be nice to know what attribute of the mushroom is a predominant factor in determining whether a mushroom is edible or poisonous. SVC with the radial basis function unfortunately, can't be used to determine the feature importance because the seperating plane and input features are in different spaces. I then decided to train an interpretable model--The Random Forest

# Cross-validation & Model fitting using Random Forest

```python
# Number of trees in Random Forest
rf_n_estimators = [int(x) for x in np.linspace(200, 1000, 5)]
rf_n_estimators.append(1500)
rf_n_estimators.append(2000)

# Maximum number of levels in tree
rf_max_depth = [int(x) for x in np.linspace(5, 55, 11)]
# Add the default as a possible value
rf_max_depth.append(None)

# Number of features to consider at every split
rf_max_features = ['auto', 'sqrt', 'log2']

# Criterion to split on
rf_criterion = ['gini', 'entropy']

# Minimum number of samples required to split a node
rf_min_samples_split = [int(x) for x in np.linspace(2, 10, 9)]

# Minimum decrease in impurity required for split to happen
rf_min_impurity_decrease = [0.0, 0.05, 0.1]

# Create the grid
rf_grid = {'n_estimators': rf_n_estimators,
               'max_depth': rf_max_depth,
               'max_features': rf_max_features,
               'criterion': rf_criterion,
               'min_samples_split': rf_min_samples_split,
               'min_impurity_decrease': rf_min_impurity_decrease}
```

```python
# Create the model to be tuned
from sklearn.ensemble import RandomForestClassifier
n_iter = 100 ### Setting low for now, so it will go fast.
folds = 4
rf_base = RandomForestClassifier()

# Create the random search Random Forest
rf_random = RandomizedSearchCV(estimator = rf_base, param_distributions = rf_grid, scoring="neg_mean_squared_error",
                               n_iter = n_iter, cv = folds, verbose = 2, random_state = 42, 
                               n_jobs = -1)

# Fit the random search model
rf_random.fit(X_train, Y_train)

# View the best parameters from the random search
print(rf_random.best_params_)

rf_model = rf_random.best_estimator_
rf_model.fit(X_train, Y_train)
```

# Model Metrics

```python
# model metrics using f1_score
Y_pred_rf=rf_model.predict(X_test)
f1_score(Y_test, Y_pred_rf)
```

The Random forest also gives a perfect prediction, same as the SVC. The Random forest also has the added advantage of being very interpretable and can thus be used to determine what mushroom attributes are predominat in determining whether a mushroom is edible or poisonous.

# Feature importance using Random forest

```python
rf_model.feature_importances_
```

```python
%matplotlib inline
sorted_idx = rf_model.feature_importances_.argsort()
plt.barh(X_df.columns[sorted_idx], rf_model.feature_importances_[sorted_idx])
plt.xlabel("Random Forest Feature Importance")
plt.show
```

It appears the Spore print color and the Mushroom Odor are the Predominant factor in determining whether a mushroom is poisonous or not. The model complexity can be reduced by removing the veil color, gill attachment and veil type. As their importance are negligible according to the plot

# Conclusion

- Both the SVC and Random forest performed excellently in determining whether a mushroom is edible or  poisonous.
- When selecting or purchasing mushrooms, the Spore print color and the Mushroom Odor are the two most important attribute to pay attention to.
